```python
#nfl_df[(nfl_df['Season'] >= 1990) & (nfl_df['Season'] < 2016)].groupby('Season')['PF'].sum().plot(linewidth=3)
lines = ['a']

for i in range(0,len(nfl_df['Division'].unique())):
    x = nfl_df['Division'].unique()[i]
    nfl_df[(nfl_df['Season'] >= 2000) 
           & (nfl_df['Season'] < 2016)
          & (nfl_df['Division'] == x)].groupby('Season')['PF'].sum().plot(linewidth=3, label=x)

plt.title('Total points scored per season (by team)',size=10)
plt.ylabel('Points scored')
print(lines)
plt.legend(bbox_to_anchor=(1.5,1))
```

```python
nfl_df_presPartyGroup = nfl_df[(nfl_df['Season'] < 2016) & (nfl_df['Season'] >= 1990)].groupby(['Team', 'PresParty']).mean()
nfl_df_presPartyGroup_division = nfl_df[(nfl_df['Season'] < 2016) & (nfl_df['Season'] >= 1990)].groupby(['Division', 'PresParty']).mean()


nfl_df_presPartyGroup.reset_index(inplace=True)
nfl_df_presPartyGroup_division.reset_index(inplace=True)

    
nfl_presParty_pivoted_PD = nfl_df_presPartyGroup\
    .pivot(index='Team', columns='PresParty', values='PD')

nfl_presParty_pivoted_wins = nfl_df_presPartyGroup\
    .pivot(index='Team', columns='PresParty', values='Wins')
    
nfl_presParty_division_pivoted_PD = nfl_df_presPartyGroup_division\
    .pivot(index='Division', columns='PresParty', values='PD')

nfl_presParty_division_pivoted_wins = nfl_df_presPartyGroup_division\
    .pivot(index='Division', columns='PresParty', values='Wins')
```

```python
ax = nfl_presParty_pivoted_PD.plot.bar(stacked=False,figsize=(15,8), color=['b', 'r'])
plt.title('Mean point differential, separated by political party of current US president')
plt.ylabel('Point differential (points scored - points against)')

ax = nfl_presParty_pivoted_wins.plot.bar(stacked=False,figsize=(15,8), color=['b', 'r'])
plt.title('Mean wins per season, separated by political party of current US president')
plt.ylabel('# wins')
```
