```python
import urllib.request
response = urllib.request.urlopen('http://nflarrest.com/api/v1/team/timeline/CHI')
html = response.read()
html_str = html.decode('ascii')
html_list = eval(html_str)
result_dict = {}
for item in html_list:
    #print(item)
    if item['Year'] in result_dict:
        result_dict[item['Year']] += int(item['arrest_count'])
    else:
        result_dict[item['Year']] = 1

print(result_dict)
```
